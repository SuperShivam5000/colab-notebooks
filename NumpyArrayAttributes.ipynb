```python
import numpy as np
```

```python
#shape
a=np.array([[1,2,3],[4,5,6]])
print(a.shape)
```

```python
#reshaping
a=np.array([[1,2,3],[4,5,6]])
a.shape=(3,2)
print(a)
```

```python
#ndim
a=np.arange(24)
print(a)
print(a.ndim)
```

```python
b=a.reshape(2,4,3) #3 dimensions
print(b)
```

```python
#itemsize
x=np.array([1,2,3,4,5], dtype=np.int8)
print(x)
print(x.itemsize) #size of item in bytes
```

```python
x=np.array([1,2,3,4,5], dtype=np.float32)
print(x)
print(x.itemsize)
```
