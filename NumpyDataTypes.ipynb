```python
import numpy as np
```

```python
dt = np.dtype(np.int32)
print(dt)
```

```python
#int8, int16, int32, int64 are equivalent to i1,i2,i4,i8

dt=np.dtype('i1')
print(dt)

dt=np.dtype('i2')
print(dt)

dt=np.dtype('i4')
print(dt)

dt=np.dtype('i8')
print(dt)
```

```python
#using endian notation
dt=np.dtype('>i4')
print(dt)
```

```python
dt=np.dtype([('age',np.int8)])
print(dt)
```

```python
#applying to n-dimensional array
dt=np.dtype([('age',np.int8)])
a=np.array([(10,),(20,),(30,)], dtype = dt)
print(a)
```
